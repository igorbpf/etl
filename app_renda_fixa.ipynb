```python
from pprint import pprint
import json
import requests

pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
url = "https://api2.apprendafixa.com.br/vn/get_featured_investments"
```

```python
data = {'tp_q': 3, 'idx': [], 'tipo': [], 'corretora': [], 'emissor': []}
```

```python
json.dumps(data)
```

```python
headers = {'user-agent': 'Mozilla/5.0'}
```

```python
resp = requests.post(url, data=json.dumps(data))
```

```python
resp.status_code
```

```python
import pandas as pd
import io
```

```python
df = pd.DataFrame(resp.json())
```

```python
df.head()
```

```python
def parse_id(val):
    return val['$oid']
```

```python
df['_id'] = df['_id'].apply(parse_id)

df.head()
```
